### Objectives
- Load and explore the dataset.

- Train a Naive Bayes classifier.
- Evaluate the model using performance metrics.

### About Naive Bayes
Naive Bayes is a probabilistic machine learning algorithm based on Bayes' Theorem. It assumes that all input features are independent of each other, making it simple, fast, and highly effective for text classification tasks such as spam detection and sentiment analysis.


```python
import numpy as np
import pandas as pd
import sklearn
train_docs = pd.read_csv('example_train.csv') 
train_docs
```

## Dataset Loading
This code imports the dataset into a Pandas DataFrame so it can be explored and processed.

```python
train_docs['Class'] = train_docs.Class.map({'cinema':0, 'education':1})
train_docs
```

```python
train_array = train_docs.values

X_train = train_array[:,0]
y_train = train_array[:,1]
y_train = y_train.astype('int') 

print("X_train")
print(X_train)
print("y_train")
print(y_train)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

## Feature Extraction
Text data is converted into numerical vectors using vectorization techniques so machine learning algorithms can process it.

```python
vec = CountVectorizer()
```

## Feature Extraction
Text data is converted into numerical vectors using vectorization techniques so machine learning algorithms can process it.

```python
vec.fit(X_train)
vec.vocabulary_
```

```python
vec = CountVectorizer(stop_words='english')
vec.fit(X_train)
vec.vocabulary_
```

## Feature Extraction
Text data is converted into numerical vectors using vectorization techniques so machine learning algorithms can process it.

```python
print(vec.get_feature_names_out())
print(len(vec.get_feature_names_out()))
```

```python
X_transformed = vec.transform(X_train)
X_transformed
```

```python
print(X_transformed)
```

```python
X_transformed.toarray()
```

```python
pd.DataFrame(X_transformed.toarray(), 
             columns=vec.get_feature_names_out())
```

```python
test_docs = pd.read_csv('example_test.csv') 
test_docs
```

## Dataset Loading
This code imports the dataset into a Pandas DataFrame so it can be explored and processed.

```python
test_docs['Class'] = test_docs.Class.map({'cinema':0, 'education':1})
test_docs
```

```python
test_numpy_array = test_docs.values

X_test = test_numpy_array[:,0]
y_test = test_numpy_array[:,1]

print("X_test")
print(X_test)
print("y_test")
print(y_test)
```

```python
X_test_transformed = vec.transform(X_test)
X_test_transformed
```

```python
X_test = X_test_transformed.toarray()
X_test
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(X_transformed, y_train)
proba = mnb.predict_proba(X_test)
```

## Model Training
The Naive Bayes classifier is trained using the training data to learn patterns between features and target labels.

```python
print("probability of test document belonging to class CINEMA", proba[:,0])
print("probability of test document belonging to class EDUCATION", proba[:,1])
```

```python
from sklearn.naive_bayes import BernoulliNB

# instantiating bernoulli NB class
bnb=BernoulliNB()

# fitting the model
bnb.fit(X_transformed, y_train)

bnb.predict_proba(X_test)
prob_bnb = bnb.predict_proba(X_test)
prob_bnb
```
